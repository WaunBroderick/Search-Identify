```python
from matplotlib import image, patches, pyplot as plt
from darknetpy.detector import Detector
from shutil import copyfile
import cv2

import os
import io


from IPython.display import display
```

```python
detector = Detector(
    '/home/csec/Desktop/darknet-master/cfg/flag.data',
    '/home/csec/Desktop/darknet-master/cfg/flag.cfg',
    '/home/csec/Desktop/darknet-master/flag_800.weights'
)
```

```python
directory = '/home/csec/Documents/Python/Notebooks/flagObjectFinder/tests/'
cutOff = 0.5


if not os.path.exists(directory + "/positive"):
    os.makedirs(directory + "/positive")
    
positiveDir = (directory + 'positive/')

for filename in os.listdir(directory):
            if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
                boxes = detector.detect(directory + filename)
                for i in range(len(boxes)):
                    i += 1
                    dictBox = dict(boxes[i-1])
                    if dictBox['class'] == 'flag' and dictBox['prob'] > cutOff:
                        print(boxes)
                        print(filename)
                        print(i)
                        copyfile(directory + filename, positiveDir + filename)
                        img = cv2.imread(positiveDir + filename)
                        crop_img = img[int(dictBox['top']):int(dictBox['left']), int(dictBox['bottom']):int(dictBox['right'])]
                        cv2.imwrite(positiveDir + filename ,crop_img)
```

```python
fig, ax = plt.subplots(1)
ax.imshow(image.imread('/home/csec/Desktop/darknet-master/data/flag2.jpg'))

colors = ['r', 'b', 'y']

for i, box in enumerate(boxes):
    l = box['left']
    t = box['top']
    b = box['bottom']
    r = box['right']
    c = box['class']
    color = colors[i % len(colors)]
    
    rect = patches.Rectangle(
        (l, t),
        r - l,
        b - t,
        linewidth = 1,
        edgecolor = color,
        facecolor = 'none' 
    )
    
    ax.text(1, t, c, fontsize = 12, bbox = {'facecolor': color, 'pad': 2, 'ec': color })
    ax.add_patch(rect)
    
plt.show()
```
